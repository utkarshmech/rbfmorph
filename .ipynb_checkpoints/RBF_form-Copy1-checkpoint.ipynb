```python
nx, ny, nz = (10, 10, 10)
mesh = np.zeros((nx * ny * nz, 3))

xv = np.linspace(0, 1, nx)
yv = np.linspace(0, 1, ny)
zv = np.linspace(0, 1, nz)
z_1, y_1, x_1 = np.meshgrid(zv, yv, xv)

# mesh = np.array([x.ravel(), y.ravel(), z.ravel()])
# mesh = mesh.T
```

```python
num_coor=len(x_1.flatten())
coor=np.zeros((num_coor,3))
coor[:,0]=x_1.flatten()
coor[:,1]=y_1.flatten()
coor[:,2]=z_1.flatten()
```

```python
fig = pl.figure(1)
ax = fig.add_subplot(111, projection='3d')
ax.scatter(coor[:, 0], coor[:, 1], coor[:, 2], c='blue', marker='o')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
pl.show()
```

```python
# inputs
N=8
ctr_pts=np.zeros((N,3))
ctr_pts[0,:]=(-0.1,   0.0,   0.0)
ctr_pts[1,:]= (0.0 ,  0.0 ,  1.0)
ctr_pts[2,:]= (0.0  , 1.0  , 0.0)
ctr_pts[3,:]= (1.0   ,0.0   ,0.0)
ctr_pts[4,:]= (0.0,   1     ,1.0)
ctr_pts[5,:]= (1.0 ,  0.0  , 1.0)
ctr_pts[6,:]= (1.0  , 1.0 ,    0)
ctr_pts[7,:]= (1.0   ,1.0,   1.0)
g=np.zeros((N,3))
g[0,:]= (0.1,  0.2,   0.3)
g[1,:]= (0.0 ,  0.0 ,  1.0)
g[2,:]= (0.0  , 1.0  , 0.0)
g[3,:]= (1.0   ,0.0   ,0.0)
g[4,:]= (0.0,   .8     ,1.0)
g[5,:]= (1.0 ,  0.0  , 1.0)
g[6,:]= (1.0  , 1.0 ,  0)
g[7,:]= (1.2   ,1.2,   1.2)
```

```python
N=8
P=np.zeros((N,4))
for i in range(N):
    for j in range(4):
        if (j==0): P[i,j]=1
        else: P[i,j]=ctr_pts[i,j-1]
# known displacements
```

```python
P
```

```python
def rd(c1, c2):
    return np.sqrt((c1[0]-c2[0])**2+(c1[1]-c2[1])**2+(c1[2]-c2[2])**2)
#rbf as global support spline type
def rbf(r):
    return np.exp(-r**2)
```

```python
def M(coor,ctr_pts):
    M_m=np.zeros((N,N))
    for i in range(N):
        for j in range(N):
            M_m[i,j]=rbf(rd(ctr_pts[i],ctr_pts[j]))
    return M_m
def lhs_(P,M_m):
    lhs=np.zeros((N+4,N+4))
    P_t=np.transpose(P)
    for i in range(N+4):
        for j in range(N+4):
            if (i<N and j<N): lhs[i,j]=M_m[i,j]
            elif (i<N and j>(N-1)): lhs[i, j]=P[i,j-N]
            elif (i>(N-1) and j<N):lhs[i,j]= P_t[i-N,j]
    return lhs
def sol_yb(lhs, g):
    rhs=np.zeros((N+4,3))
    for i in range(N):
        for j in range(3):    
            rhs[i,j]=g[i,j]
    
    inv_lhs=np.linalg.inv(lhs)
    yb= inv_lhs @ rhs
    gmma=np.zeros((N,3))
    beta=np.zeros((4,3))
    for i in range(N+4):
        if (i<N): gmma[i,:]=yb[i,:]
        else : beta[i-N,:]=yb[i,:]
    return (gmma, beta)
                
```

```python
def s_x(ctr_pts, coor, beta, gmma,deg):
    s_x=np.zeros(3)
    for i in range(3):
        for j in range(N):
            s_x[i]=s_x[i]+gmma[j,i]*rbf(rd(coor, ctr_pts[j]),deg)
        s_x[i]=s_x[i]+beta[0,i]*1 + beta[1,i]*coor[0] + coor[1]*beta[2,i] + coor[2]*beta[3,i]
    return s_x
```

```python
M_m=M(coor,ctr_pts)
# M_m
```

```python
lhs=lhs_(P,M_m)
# lhs
```

```python
gmma, beta=sol_yb(lhs, g)
```

```python
coor_new=np.zeros((num_coor,3))
```

```python
cor_t=(1.0,1.0,1.0)
s_x(ctr_pts, cor_t, beta, gmma,deg)
```

```python
for i in range(num_coor):
    coor_new[i,:]=coor[i,:]+s_x(ctr_pts, coor[i], beta, gmma,deg)
```
